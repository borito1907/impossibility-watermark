```python
import matplotlib.pyplot as plt
import pandas as pd

prompt_name = "pen"

# Load the CSV file into a DataFrame
df = pd.read_csv(f'./{prompt_name}.csv')

# Ensure the 'watermark_score' is treated as a float
df['watermark_score'] = df['watermark_score'].astype(float)

# Plotting the evolution of the watermark_score (z-score)
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['watermark_score'], marker='o', linestyle='-', color='b', label='Z-Score (Watermark Score)')

plt.title(f'Evolution of Z-Score (Watermark Score) for {prompt_name}')
plt.xlabel('Attack Evolution (Row Number)')
plt.ylabel('Z-Score (Watermark Score)')
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

## Analysis

```python
pen_attack_df = pd.read_csv('./pen.csv')
```

```python
current_text = pen_attack_df.iloc[-1]['current_text']
mutated_text = pen_attack_df.iloc[-1]['mutated_text']
```

```python
from typing import *
from nltk.tokenize import sent_tokenize
import re

def handle_bullet_points(sentences: List[str]) -> List[str]:
    new_sentences = []
    digit_pattern = re.compile(r'^\*?\*?\d+\.$')
    i = 0
    num_sentences = len(sentences)
    if num_sentences == 0:
        return sentences
    # log.info(f"Num sentences: {num_sentences}")
    while i < num_sentences - 1:
        if digit_pattern.match(sentences[i].strip()):
            modified_sentence = f"{sentences[i].strip()} {sentences[i + 1]}"
            new_sentences.append(modified_sentence)
            # log.info(f"Adding {modified_sentence}")
            i += 1  # Skip the next element as it's already added
        else:
            new_sentences.append(sentences[i])
        i += 1
        # log.info(f"i={i}")
    # Add the last sentence as well, if we don't want to skip it
    if i == num_sentences - 1:
        new_sentences.append(sentences[-1])
    # log.info(f"Sentences: {new_sentences}")
    return new_sentences

def tokenize_sentences(text: str) -> List[str]:
    sentences = sent_tokenize(text)
    processed_sentences = handle_bullet_points(sentences)
    return processed_sentences
```

```python
tokenize_sentences(current_text)
```

```python
tokenize_sentences(mutated_text)
```
