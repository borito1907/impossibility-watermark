```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from utils import count_words
```

```python
entropy = 4
output = 1
attack_id = "edit_1"

csv_file_directory = f"results/stationary_distribution/robustness_analysis/entropy_{entropy}/"

csv_filename = f"output_{output}/corpuses/attack_{attack_id}.csv"
csv_file_path = os.path.join(csv_file_directory, csv_filename)

df = pd.read_csv(csv_file_path)

# Calculate the ratio of mutated_text_len to current_text_len
df['ratio'] = df['mutated_text_len'] / df['current_text_len']

# Filter out rows where the ratio is less than 0.8
filtered_df = df[df['ratio'] >= 0.8]

# Plot the ratio as a function of step_num for filtered data
plt.figure(figsize=(10, 6))
plt.plot(filtered_df['step_num'], filtered_df['ratio'], marker='o', linestyle='-', color='b')
plt.title('Ratio of Mutated Text Length to Current Text Length over Steps (Ratio >= 0.8)')
plt.xlabel('Step Number')
plt.ylabel('Ratio (Mutated Text Length / Current Text Length)')
plt.grid(True)
plt.show()
```
