## UMD Generations

```python
import pandas as pd

regen_1 = pd.read_csv('/local1/borito1907/impossibility-watermark/llama31_dev_gens/umd_dev_llama31_regen/watermarked_texts.csv')

regen_2 = pd.read_csv('/local1/borito1907/impossibility-watermark/llama31_dev_gens/umd_dev_llama31_regen_2/watermarked_texts.csv')

polished = pd.read_csv('/local1/borito1907/impossibility-watermark/llama31_dev_gens/umd_first_round_polished.csv')

combined_df = pd.concat([regen_1, regen_2, polished], axis=0)
combined_df = combined_df.sort_values('zscore', ascending=False).drop_duplicates('id')
combined_df
```

```python
combined_high = combined_df[combined_df['zscore'] > 3]
combined_high
```
