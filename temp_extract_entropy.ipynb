```python
import re

def extract_entropy_from_logfile(logfile_path):
    entropy_values = []
    start_capturing = False

    # Regular expression to match entropy lines
    entropy_pattern = re.compile(
        r"\[\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2},\d{3}\]\[watermarkers\.adaptive\]\[INFO\] - Entropy: ([\d\.]+)"
    )

    # Define start and stop markers
    start_marker = "[2024-10-23 14:06:14,551][__main__][INFO] - Prompt: Create a psychological portrait of a person who values their time 1000000000usd per second"
    stop_marker = "[2024-10-23 14:42:01,398][__main__][INFO] - Score: 95.49179673194885"

    with open(logfile_path, "r") as logfile:
        for line in logfile:
            line = line.strip()
            
            if line == start_marker:
                start_capturing = True
                continue  # Move to the next line after the start marker

            if line == stop_marker:
                break  # Stop reading when the stop marker is found

            if start_capturing:
                # Try to match the entropy pattern
                match = entropy_pattern.match(line)
                if match:
                    # Extract the entropy value and add it to the list
                    entropy_value = float(match.group(1))
                    entropy_values.append(entropy_value)

    return entropy_values

# Usage example
logfile_path = "/data2/borito1907/impossibility-watermark/10_23_adaptive_dev_gen_clean.log"
entropy_list = extract_entropy_from_logfile(logfile_path)
print("Extracted Entropy Values:", entropy_list)
print("Length of the Extracted Entropy Values:", len(entropy_list))
```

```python
def compare_lists_from_end(list1, list2, threshold=0.10):
    # Determine the minimum length
    min_length = min(len(list1), len(list2))
    
    differences = []

    # Iterate from the end of both lists, up to the minimum length
    for i in range(1, min_length + 1):
        diff = abs(list1[-i] - list2[-i])
        if diff > threshold:
            differences.append((i, list1[-i], list2[-i]))

    return differences
```
