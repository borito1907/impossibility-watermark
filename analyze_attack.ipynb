```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "./attack_1705194355.csv" 

df = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['step_num'], df['z_score'], marker='o')
plt.title('Z-Score as a Function of Step Number')
plt.xlabel('Step Number')
plt.ylabel('Z-Score')
plt.grid(True)
plt.show()
```

```python
quality_score_counts = df['quality_score'].value_counts().sort_index()

# Display the table of quality scores and their counts
quality_score_counts_df = quality_score_counts.reset_index()
quality_score_counts_df.columns = ['Quality Score', 'Number of Rows']
quality_score_counts_df
```
